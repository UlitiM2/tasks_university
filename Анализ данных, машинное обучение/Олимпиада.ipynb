```python
import pandas as pd
import numpy as np
```

```python
sample = pd.read_csv("Property_rent_merged.csv")
```

```python
sample
```

```python
sample['floor'].unique()
```

```python
sample.info()
```

```python
sample[['floor_number', 'total_floors']] = sample['floor'].str.split('/', expand=True)
sample['floor_number'] = pd.to_numeric(sample['floor_number'], errors='coerce')
sample['total_floors'] = pd.to_numeric(sample['total_floors'], errors='coerce')
sample['is_last_floor'] = (sample['floor_number'] == sample['total_floors']).astype(int)
sample['floor_number'] = sample['floor_number'].fillna(sample['floor_number'].mean()).astype(int)
sample['total_floors'] = sample['total_floors'].fillna(sample['total_floors'].mean()).astype(int)
```

```python
sample[sample['is_last_floor'] == 1]
```

```python
cols_to_keep = [
    'property_size',
    'ceiling_height',
    'price_per_month',
    'is_access_system',
    'is_elevator',
    'is_crane_equipment',
    'is_company',
    'floor_number',
    'is_last_floor',
    'description_y'
]
sample = sample[cols_to_keep]
```

```python
dummies = pd.get_dummies(sample['description_y'], drop_first=True)
sample = pd.concat([sample.drop(columns=['description_y']), dummies], axis=1)
```

```python
sample['is_access_system'] = sample['is_access_system'].astype(int)
sample['is_crane_equipment'] = sample['is_crane_equipment'].astype(int)
sample['is_company'] = sample['is_company'].astype(int)
sample['is_elevator'] = sample['is_elevator'].astype(int)
```

```python
sample
```

```python
for col in sample.select_dtypes(include=[bool]):
    sample[col] = sample[col].astype(int)
```

```python
sample
```

```python
mean_ceiling_height = sample['ceiling_height'].mean()
sample['ceiling_height'] = sample['ceiling_height'].fillna(mean_ceiling_height)
sample
```

```python
sample = sample.dropna()
```

```python
import pandas as pd
import numpy as np

def remove_outliers(df):
    df_copy = df.copy()


    numeric_columns = df.select_dtypes(include=np.number).columns
    for col in numeric_columns:

        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1

        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr


        outliers_index = df[(df[col] < lower_bound) | (df[col] > upper_bound)].index

        index_to_drop = [idx for idx in outliers_index if idx in df_copy.index]

        df_copy.drop(index_to_drop, inplace=True)

    return df_copy


cleaned_df = remove_outliers(sample)
print(cleaned_df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score


X = cleaned_df.drop(columns=['price_per_month'])
y = cleaned_df['price_per_month']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=21)

model = LinearRegression()
model.fit(X_train, y_train)


y_pred = model.predict(X_test)

r2 = r2_score(y_test, y_pred)
rounded_r2 = round(r2, 3)

print(rounded_r2)
```
